```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
company_sales = pd.read_csv('company_sales_data - company_sales_data.csv', index_col = 'month_number')
```

```python
company_sales
```

# Total profit 

```python
company_sales['total_profit'].plot(figsize = (16,8))
```

# Product sales 

```python
for x in company_sales.drop(['total_profit', 'total_units'], axis=1):
    company_sales[x].plot(label = x, figsize = (20,10))

plt.legend();
```

# Total sales data pie chart 

```python
sizes = [0, 0, 0, 0, 0, 0]
i = 0
for x in company_sales.drop(['total_profit', 'total_units'], axis=1): 
    sizes[i] = sum(company_sales[x])
    i+=1
```

```python
sizes
```

```python
plt.pie(sizes, labels = company_sales.drop(['total_profit', 'total_units'], axis=1).columns)
plt.tight_layout()
plt.show()
```
